```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
```

```python
#G reservoir
k_dist_flow = pd.read_csv('../data/rawdata/52.csv')
g_resv_flow = pd.read_csv('../data/rawdata/53.csv')
g_resv_lvl1 = pd.read_csv('../data/rawdata/54.csv')
g_resv_lvl2 = pd.read_csv('../data/rawdata/55.csv')
k_dist_pres = pd.read_csv('../data/rawdata/56.csv')
```

```python
k_dist_flow.columns = ['id','time', 'k_dist_flow','drop']
g_resv_flow.columns = ['id', 'time', 'g_resv_flow', 'drop']
g_resv_lvl1.columns = ['id', 'time', 'g_resv_lvl1', 'drop']
g_resv_lvl2.columns = ['id', 'time', 'g_resv_lvl2', 'drop']
k_dist_pres.columns = ['id', 'time', 'k_dist_pres', 'drop']
```

```python
del k_dist_flow['id']
del k_dist_flow['drop']
del g_resv_flow['id']
del g_resv_flow['drop']
del g_resv_lvl1['id']
del g_resv_lvl1['drop']
del g_resv_lvl2['id']
del g_resv_lvl2['drop']
del k_dist_pres['id']
del k_dist_pres['drop']
```

```python
time = k_dist_flow[:1440]['time']
x = pd.to_datetime(time)

fig, ax = plt.subplots(figsize=(15,5))

#=========k_dist_flow
#=========k_dist_flow
#=========k_dist_flow
y = k_dist_flow[:1440]['k_dist_flow']

ax.plot(x, y, label='k_dist', color='blue')


#=========g_resv_flow
#=========g_resv_flow
#=========g_resv_flow
z = g_resv_flow[:1440]['g_resv_flow']

ax.plot(x, z, label='g_resv', color='red')

#=========misc config
#=========misc config
#=========misc config
ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.set_xlabel('2023/01/01')
ax.set_ylabel('flow at dist K, m^3/hour')
ax.legend()

plt.xticks(rotation=45)

plt.show()
```
