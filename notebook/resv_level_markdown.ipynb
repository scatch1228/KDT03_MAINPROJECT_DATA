**Hazen-Williams Equation**

$$h_f = \frac{10.67 \cdot L \cdot Q^2}{C^{1.85}\cdot D^{4.87}}$$

where 

$h_f$ head (pressure) loss,
$L$ length of pipe,
$D$ diam of pipe,
$Q$ flow,
$C$ roughness coeff


**토출압력, 송수압력, 송수량으로 펌프가동여부를 확인하기 어려운 이유**
- 일단 송수량을 늘리기 위해 펌프를 가동하는 것은 일부 맞다. 
- 허나 압력과 송수량은 단순히 펌프가동여부의 영향만 받는 것은 아니다.
- 압력은 송수량과 반비례한다 (정확히는 압력손실이 송수량제곱에 비례한다). 때문에 펌프를 가동해서 송수량이 늘어날 경우에는 압력이 줄어든다. 
- 송수량은 펌프를 가동할 때도 늘어나지만, 그렇지 않을 때도 늘어난다. 여러 배수지 벨브가 열려 물이 흘러들어갈 길이 갑자기 넓어지면 송수량이 급격하게 증가한다. 이에 따라 압력도 급격히 감소한다. 높은 송수량을 유지하기 위해 펌프가 가동되지만 토출압력은 낮은 경우이다.
- 반대의 경우도 있다. 배수지 벨브가 대부분 닫혀있어 물이 갈 곳이 없어지면 송수량이 급격히 감소한다. 이때 압력이 급격하게 증가하는데, 펌프는 이 압력에 맞추기 위해 가동된다. 송수량은 없으나 높은 토출압력으로 펌프는 가동되는 경우이다. 펌프는 단순히 송수량을 증가시키기위해 가동되는 것이 아니라는 의미이다.
- 또한 일반적으로는 송수량을 증가시키기 위해 펌프를 가동하나, 송수량을 정확한 값만큼 증가시킬 수는 없다. 증가는 시키는데 얼마나 증가시킬지 모른다는 것이다.
- 때문에 절대적인 압력, 송수량만으로 펌프가동여부를 판단하기는 굉장히 어렵다.
- 물의 흐름은 펌프보다는 물이 흐를 수 있는 길이 얼마나 큰 가에 더 큰 영향을 받는다. 펌프는 물이 올바른 방향으로 흐르도록 도와주는 보조역할을 할 뿐이다.  

**보완법**
- 4대의 펌프가 동일한 성능을 지녔다고 가정했을때, 송수량과 토출압력이 동시에 급격하게 변한 후 일정수준을 유지한다면 펌프1대가 추가가동/정지 했다고 가정할 수 있다.
- 유체역학을 무시하고 단순히 데이터로만 학습을 시킬 수도 있다. 프로젝트의 실용성은 떨어지겠지만, 데이터 제약사항이 명확한 상황에서는 데이터분석 및 코딩 역량을 보여주는 것으로는 충분하다고 판단된다.
